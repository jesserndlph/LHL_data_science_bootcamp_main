```python
import pandas as pd
import json
```

```python
df = pd.read_json('world_bank_projects.json')
df.head()
```

```python
df1 = df['countryname'].value_counts() #1
df1.head(10)
```

```python
import json
#load json object
with open('world_bank_projects.json') as f:
    nested_json = json.load(f)
print(type(nested_json))
```

```python
from pandas.io.json import json_normalize  
sector = json_normalize(nested_json,record_path ='sector')
sector.head()
```

```python
df1 = df['sector'].value_counts() #2
df1.head(10)
```

```python
df3 = df['mjtheme_namecode'].value_counts() #2
df3.head(10)
```

```python
df4 = json_normalize(df3,record_path ='name')
df4.head()
```

```python
from pandas.io.json import json_normalize  
theme = json_normalize(nested_json,record_path ='mjtheme_namecode')
theme.head(20)
```
