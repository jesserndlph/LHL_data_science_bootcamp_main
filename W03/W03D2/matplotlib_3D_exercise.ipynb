## Matplotlib 3D Assignment

```python
# import mplot3d
from mpl_toolkits import mplot3d

# import numpy
import numpy as np

# import matplotlib
import matplotlib.pyplot as plt
```

------------

#### Task 1: Create wireframe of hyperbolic paraboloid on the domain <-10,10>X<-10,10>

* hyperbolic paraboloid equation is: z = x*x - y*y

```python
# wireframe
def f (x,y):
    return (x**2)-(y**2)
```

```python
x = np.linspace(-10, 10, 30)
y = np.linspace(-10, 10, 30)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

#### Task 2: Create surface plot of eliptical paraboloid on the domain <-10,10>X<-10,10>

* eliptical paraboloid : z = x*x + y*y

```python
# surface plot
def q (x,y):
    return (x**2)+(y**2)
x = np.linspace(-10, 10, 30)
y = np.linspace(-10, 10, 30)
X, Y = np.meshgrid(x, y)
Z = q(X, Y)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```
