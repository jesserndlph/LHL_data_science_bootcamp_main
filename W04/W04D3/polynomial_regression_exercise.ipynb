#### Task 1

Write a function psi(x,m) defined as $\psi(x)=[1, x, x ^2, ..., x^m]$

```python
def psi(x,m):
    psilist = [1]
    z=1
    for i in range(1,m+1):
        psilist.append(x**z)
        z+=1
    return psilist
```

#### Task 2

Write a function $h(x,\alpha, m)$ defined as $h(x,\alpha, m)=\alpha^T\psi(x, m)$

```python
def h(x,a,m):
    return np.dot([a.T, psi(x,m)])
```
