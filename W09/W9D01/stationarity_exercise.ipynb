* generate a stationary time-series

```python
# generate stationary time series
def stationary_time_series(x):
    # create a list to store the time series
    y = []
    # iterate through the time series
    for i in range(len(x)):
        # if the time series is the first element
        if i == 0:
            # append the first element
            y.append(x[i])
        # if the time series is not the first element
        else:
            # append the difference between the current element and the previous element
            y.append(x[i] - x[i - 1])
    # return the time series
    return y
```

* plot the stationary series generated above

```python
x_stationary = stationary_time_series(x)
```

* Create a function `'ADF_Stationarity_Test'` with the input parameters `timeseries` and `significance_level` which: 
    * prints the adfuller test results
    * prints the conclusion stating whether timeseries is stationary based on the siginificance level (input parameter of function)


```python
def ADF_Stationarity_Test(timeseries, significance_level=0.05):
    # calculate the mean
    mean = sum(timeseries) / len(timeseries)
    # calculate the variance
    variance = sum([(x - mean)**2 for x in timeseries]) / len(timeseries)
    # calculate the standard deviation
    std_dev = variance**0.5
    # calculate the skewness
    skewness = sum([(x - mean)**3 for x in timeseries]) / len(timeseries)**1.5
    # calculate the kurtosis
    kurtosis = sum([(x - mean)**4 for x in timeseries]) / len(timeseries)**2
    # calculate the z-score
    z = (mean - 0) / std_dev
    # calculate the p-value
    p = 1 - (1 - (1 - significance_level)**(1 / len(timeseries)))**(1 / len(timeseries))
    # calculate the critical value
    critical_value = std_dev * (p**(1 / len(timeseries)))
    # return the z-score, p-value, critical value
    return z, p, critical_value
```
